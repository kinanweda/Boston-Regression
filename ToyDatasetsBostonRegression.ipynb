**2. Load Data Boston**

```python
df = pd.DataFrame(
    dataBoston['data'],
    columns = dataBoston['feature_names']
)
df['Price'] = dataBoston['target']
df
```

```python
corr = df.corr()
plt.imshow(corr, cmap='BuPu_r')
plt.colorbar()
plt.xticks(np.arange(14), list(df.columns),rotation=90)
plt.yticks(np.arange(14), list(df.columns))
plt.show()
```

```python
model= linear_model.LinearRegression()
model.fit(df[['ZN','CHAS','RM','DIS','B']],df['Price'])
model.predict(df[['ZN','CHAS','RM','DIS','B']])[0]
```
